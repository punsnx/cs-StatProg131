```python
import numpy as np
import pandas as pd
from scipy import stats
from statsmodels.stats.proportion import proportions_ztest
import cmath
```

8.1 ร้านค้าออนไลน์ขายเมล็ดพืชหายากชนิดหนึ่ง อ้างว่าเมล็ดที่ขายมีเปอร์เซ็นต์การงอกสูงกว่า 90% จากการทดลองของลูกค้ารายหนึ่งได้สั่งซื้อเมล็ดมาทั้งหมด 100 เมล็ดและพบว่ามี 14 เมล็ดที่ไม่งอก จงทดสอบคำกล่าวอ้างของร้านค้าว่าเป็นจริงหรือไม่ที่ระดับนัยสำคัญ 0.05 หากถือว่าอัตราการงอกของเมล็ดพืชดังกล่าวมีการแจกแจงแบบปกติ

```python
#แสดงสมมติฐาน พร้อมวิธีหาค่าcal ค่าp-value และการสรุปผล
# 1.สมมติฐาน H0 : P <= P0, H1 : P > P0
x = 100-14 # x we interested 100 - 14 ไม่งอก = 86 งอก
n = 100 # n observation
alpha = 0.05
P0 = 0.90
Q0 = 1-P0
p_hat = x/n
# 2.หา Z_cal
z_cal = p_hat-P0/np.sqrt((P0*Q0)/n)
z_crit = stats.norm.isf(alpha)
print(z_cal,z_crit)
# 3.หา p value
p_val = stats.norm.sf(z_cal)
print(p_val,alpha)
#สรุป
print("สรุป เนื่องจาก z_cal < z_crit และ p value > alpha จึง accept H0 reject H1")
print("ไม่ยอมรับสมมติฐานที่ว่า เมล็ดที่ขายมีเปอร์เซ็นต์การงอกสูงกว่า 90%")
```

8.2 สืบเนื่องจากข้อ 8.1 พบว่ามีร้านค้าออนไลน์อีกเจ้าซึ่งขายเมล็ดพืชหายากชนิดเดียวกันนี้ อ้างว่าเมล็ดที่ขายมีเปอร์เซนต์การงอกสูงกว่า 95% จึงสั่งซื้อมา 50 เมล็ดพบว่ามี 4 เมล็ดที่ไม่งอก จงทดสอบว่าร้านค้านี้ขายเมล็ดซึ่งมีเปอร์เซ็นต์การงอกสูงกว่าร้านค้าแรกมากกว่าหรือเท่ากับ 5% หรือไม่ ที่ระดับนัยสำคัญ 0.05

```python
#แสดงสมมติฐาน พร้อมวิธีหาค่าcal ค่าp-value และการสรุปผล
# 1.สมมติฐาน H0 : P1 - P2 >= P0, H1 : P1 - P2 < P0 ; P1 ร้าน2 P2 ร้าน1
x1 = 50-4 # x we interested 50 - 4 ไม่งอก = 46 งอก
n1 = 50 # n observation 
x2 = 100-14 # x we interested 100 - 14 ไม่งอก = 86 งอก
n2 = 100 # n observation 

P0 = 0.05
P1 = 0.95
P2 = 0.90

alpha = 0.05

p_hat1 = x1/n1
p_hat2 = x2/n2 

p_hat = (x1+x2)/(n1+n2)
q_hat = 1-p_hat
# 2.หา Z_cal
z_cal = ((p_hat1-p_hat2)-P0)/np.sqrt(p_hat * q_hat * ((1/n1)+(1/n2))) 
# ใช้ cmath.sqrt เพราะรองรับ sqrt negative ; cmath module, which supports complex numbers. 
z_crit = stats.norm.ppf(alpha)
print(z_cal,z_crit, z_cal <= z_crit)
# 3.หา p value
p_val = stats.norm.cdf(z_cal)
print(p_val,alpha, p_val <= alpha)
#สรุป
print("สรุป เนื่องจาก z_cal > z_crit และ p value > alpha จึง accept H0 reject H1")
print("ยอมรับสมมติฐานที่ว่า ร้านค้านี้ขายเมล็ดซึ่งมีเปอร์เซ็นต์การงอกสูงกว่าร้านค้าแรกมากกว่าหรือเท่ากับ 5% ")
```

8.3 จากชุดข้อมูล "titanic.csv" จงทดสอบว่าส่วนเบี่ยงเบนมาตราฐานข้อมูลอายุผู้โดยสารมีค่าน้อยกว่า 10 ปีหรือไม่ที่ระดับความเชื่อมั่น 99% หากกำหนดให้ข้อมูลอายุผู้โดยสารมีการแจกแจงแบบปกติ

```python
#แสดงสมมติฐาน พร้อมวิธีหาค่าcal ค่าp-value และการสรุปผล
# 1.สมมติฐาน H0  zigma_2 >= zigma0_2 H1 : zigma_2 < zigma0_2
df = pd.read_csv("https://raw.githubusercontent.com/ThammakornS/ProgStat/main/dataset/titanic.csv")

zigma0_2 = 10**2

n = df.Age.count()
s = df.Age.std()
s_2 = s**2

v = n-1

alpha = 0.01 # 1-0.99

# 2.หา normal distribution ใช้ chi2_cal
chi2_cal = (n-1)*s_2/zigma0_2
chi2_crit = stats.chi2.ppf(alpha,df=v)
print(chi2_cal,chi2_crit, chi2_cal <= chi2_crit)
# 3.หา p value
p_val = stats.chi2.cdf(chi2_cal,df=v)
print(p_val,alpha, p_val <= alpha)
#สรุป
print("สรุป เนื่องจาก chi2_cal > chi2_crit และ p value > alpha จึง accept H0 reject H1")
print("ยอมรับสมมติฐานที่ว่า ส่วนเบี่ยงเบนมาตราฐานข้อมูลอายุผู้โดยสารมีค่าน้อยกว่า 10 ปี")
```

8.4 จากข้อ 8.3 จงทดสอบว่ากลุ่มผู้โดยสารชั้น Pclass = 1 มีความแปรปรวนของอายุเท่ากับ กลุ่มผู้โดยสารชั้น Pclass = 2 หรือไม่ที่ระดับนัยสำคัญ 0.05

```python
#แสดงสมมติฐาน พร้อมวิธีหาค่าcal ค่าp-value และการสรุปผล

# 1.สมมติฐาน H0  zigma1_2/zigma2_2 = 1 H1 : zigma1_2/zigma2_2 != 1
df = pd.read_csv("https://raw.githubusercontent.com/ThammakornS/ProgStat/main/dataset/titanic.csv")
n1 = df[(df.Pclass == 1)].Age.count()
n2 = df[(df.Pclass == 2)].Age.count()

x1 = df[(df.Pclass == 1)].Age
x2 = df[(df.Pclass == 2)].Age

var1 = np.var(x1,ddof=1)
var2 = np.var(x2,ddof=1)

alpha = 0.05

v1 = n1-1
v2 = n2-1

# 2.หา f distribution
f_cal = var1/var2
f_crit_left = stats.f.ppf(alpha/2, dfn=v1,dfd=v2)
f_crit_right = stats.f.isf(alpha/2, dfn=v1,dfd=v2)
print(f_cal,f_crit_left,f_crit_right)
# 3.หา p value
p_val = 2*min(stats.f.cdf(f_cal, dfn=v1,dfd=v2),stats.f.sf(f_cal, dfn=v1,dfd=v2))
print(p_val,alpha)
#สรุป
print("สรุป เนื่องจาก f_cal < f_crit_left และ p value < alpha จึง reject H0 accept H1")
print("ไม่ยอมรับสมมติฐานที่ว่า กลุ่มผู้โดยสารชั้น Pclass = 1 มีความแปรปรวนของอายุเท่ากับ กลุ่มผู้โดยสารชั้น Pclass = 2 ")
```
