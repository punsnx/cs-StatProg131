จากข้อมูล titanic.csv ให้ทำการวิเคราะห์ข้อมูลเบื้องต้นเกี่ยวกับคอลัมน์ cabin  
โดยแผนภาพด้านล่างนี้เป็นตำแหน่งของ cabin ตามตัวอักษรตัวแรกของรหัส cabin ที่บันทึกในชุดข้อมูล

ให้นิสิตสร้างคอลัมน์เพิ่ม โดยคอลัมน์ที่เพิ่มมานี้ จะเก็บข้อมูลของตัวอักษรตัวแรกที่ปรากฏในคอลัมน์ Cabin เช่น  
- หากในคอลัมน์ Cabin เป็นค่า "C123" ในคอลัมน์ที่สร้างจะบันทึก "C"  
- หากในคอลัมน์ Cabin เป็นค่า "C23 C25 C27" ในคอลัมน์ที่สร้างจะบันทึก "C"
- หากในคอลัมน์ Cabin เป็นค่า "B58 B60 C2" ในคอลัมน์ที่สร้างจะบันทึก "B"
- หากในคอลัมน์ Cabin เป็นค่า "F2" ในคอลัมน์ที่สร้างจะบันทึก "F"
- หากในคอลัมน์ Cabin เป็นค่า NaN ในคอลัมน์ที่สร้างจะบันทึก NaN

จากนั้นให้นิสิตเขียนโค้ดเพื่อตอบคำถามในแต่ละข้อ

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/ThammakornS/ProgStat/main/dataset/titanic.csv")
df.head()
```

นิสิตแสดงวิธีสร้างคอลัมน์

```python
#your code
df['cabin_letter'] = df.Cabin.str.get(0) if df.Cabin.str != np.nan else np.nan
df.dropna().head()
```

1. ในแต่ละตัวอักษร Cabin มีจำนวนผู้โดยสารกี่คน และมีจำนวน NaN เท่าไหร่

```python
#your code
df.cabin_letter.value_counts(dropna=False)
```

2. ในแต่ละตัวอักษร Cabin มีจำนวนผู้โดยสารที่รอดและเสียชีวิตอย่างละเท่าไหร่ cabin ใดมีจำนวนผู้เสียชีวิตมากที่สุด

```python
#your code
# compare = pd.crosstab(df.Survived,df.cabin_letter, margins=True)
# cabin_die = df[df.Survived == 0].cabin_letter.value_counts(dropna=True)
# print("===COMPARE===\n",compare)
# print("===MAX DIED===\n",cabin_die[cabin_die == cabin_die.max()])
cabin_survive = df.groupby(["cabin_letter","Survived"])["Survived"].value_counts()
cabin_die = df[df.Survived == 0].groupby("cabin_letter")["Survived"].value_counts()
print(cabin_survive.unstack(level=1),cabin_die[cabin_die == cabin_die.max()],sep="\n")
```

3. cabin ใดมีค่า ผู้รอดชีวิต/ผู้เสียชีวิต สูงสุด

```python
#your code
# cabin_die = df[df.Survived == 0].cabin_letter.value_counts(dropna=True)
# cabin_survive = df[df.Survived == 1].cabin_letter.value_counts(dropna=True)

# print("cabin_max_die\n",cabin_die[cabin_die == cabin_die.max()])
# print("cabin_max_survive\n",cabin_survive[cabin_survive == cabin_survive.max()])
cabin_survive_die = df[df.Survived == 1].Survived / df[df.Survived == 0].Survived
cb_survive = pd.crosstab(df.cabin_letter,df.Survived)
cb_survive_div_die = cb_survive[1]/cb_survive[0]
cb_survive_div_die[ cb_survive_div_die == cb_survive_div_die.max()]
```

4. cabin ใดมีค่าเฉลี่ยราคาตั๋ว (Fare) สูงสุด

```python
#your code
cabin_mean_fare = df.groupby("cabin_letter")["Fare"].mean()
print("max_fare\n", cabin_mean_fare[cabin_mean_fare == cabin_mean_fare.max()])
```

5. จงแสดงค่าเฉลี่ยราคาตั๋ว (Fare) ของผู้รอดชีวิต และผู้เสียชีวิต ในแต่ละ cabin หากเปรียบเทียบราคาเฉลี่ยของผู้ที่รอดและเสียชีวิตจะพบว่า

```python
#your code
each_cabin_fare = df.dropna().groupby(["cabin_letter","Survived"])["Fare"].mean()
survived_fare = df.dropna().groupby(["Survived"])["Fare"].mean()
print(each_cabin_fare.unstack(level=1),survived_fare,sep="\n")
```
